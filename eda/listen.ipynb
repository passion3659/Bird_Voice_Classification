```python
import os
import psutil
import numpy as np
# import pandas as pd

import sys
sys.path.append("..")
# from utils import *

import warnings
warnings.filterwarnings("ignore")

# from matplotlib import pyplot as plt
import librosa
import IPython.display as ipd
import matplotlib.pyplot as plt
```

```python
from IPython.display import Audio

# ogg  
audio_file = "/home/workspace/dacon/01_bird_voice_classification/data/TRAIN/TRAIN_0002.ogg"

#   
Audio(audio_file)
```

```python
audio_data, sr = librosa.load(audio_file, sr=32000)
```

```python
audio_data.shape
```

```python
S = librosa.feature.melspectrogram(y=audio_data, sr=32000, n_fft=1024)

#  
log_S = librosa.power_to_db(S, ref=np.max)

# Mel  
plt.figure(figsize=(10, 4))
librosa.display.specshow(log_S, sr=32000, x_axis='time', y_axis='mel')
plt.title('Mel spectrogram')
plt.colorbar(format='%+02.0f dB')
plt.tight_layout()
plt.show()
```
